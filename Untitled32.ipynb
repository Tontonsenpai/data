```python
def select_columns(data_frame, column_names):
    new_frame = data_frame.loc[:, column_names]
    return new_frame
selected_columns = ['session_projects_added','session_comments_given', 'bugs_in_session','session_likes_given','duration_time']
newtue = select_columns(tuesday, selected_columns)

newtue

#ax = newtue.plot.bar(rot=0)
```
